For  all the titles in the 'title_cl' column of the DataFrame data.The title is split the title into individual words and 
the counter cnt is updated with the count of each word.

```python

from collections import Counter
cnt = Counter()

data = data.dropna(subset=['title_cl'])

# Finding the occurence of words in video titles
for title in data['title_cl']:
    for token in title.split():
        cnt[token] += 1

# Finding the 50 most used words in video titles
dist = cnt.most_common(8)
top_words = []
freqs = []
for word, freq in dist:
    top_words.append(word)
    freqs.append(freq)
```

top_words and freq bar plot of top 10 words

For each word in the top_words list, a new column is added to the DataFrame data. The column contains boolean values, indicating whether the corresponding title contains the particular word. The str.contains method is used to check for the presence of the word in the 'title_cl' column, and the result is stored as a new column with the same word as the column name.

```python
# Create columns for each of the top words
for word in top_words:
    data[word] = data['title_cl'].str.contains(word, case=False)

# Calculate correlation between the top word columns and the 'viewcount' column
correlations = data[top_words + ['view_count']].corr()

data.head()
```

Each row splits the 'title_cl' column into individual words and checks if any of the words are present in the top_words list. If a word from the top_words list is found in the title, a dictionary containing the title and the corresponding 'view_count' is appended to the title_view_list.

After populating the title_view_list, the code converts it into a pandas DataFrame named title_view_df using the pd.DataFrame function. Fi

```python
title_view_list = []
for i, row in data.iterrows():
    for title in row['title_cl'].split():
        if title in top_words:
            title_view_list.append({'title': title, 'view_count': row['view_count']})

# Convert the list of dictionaries to a DataFrame
title_view_df = pd.DataFrame(title_view_list)

display(title_view_df)
```

The most common tags from the 'tags' column of the DataFrame data.and their corresponding frequencies for further analysis or visualization.

```python
cnt = Counter()

data = data.dropna(subset=['tags'])

# Finding the occurence of words in video titles
for tags in data['tags']:
    for token in eval(tags):
        cnt[token] += 1

# Get the 50 most common tags
top_tags = cnt.most_common(50)
print(top_tags)
top_words = []
freqs = []
for word, freq in top_tags:
    top_words.append(word)
    freqs.append(freq)
```

create a dataframe containing view count with tags

```python
# Get view count for top 50 tags
# Assuming 'tags' is the column containing tags and 'view_count' is the column containing the view counts
tags_counter = Counter()

# Count the occurrences of each tag
for tags in data['tags']:
    for tag in eval(tags):
        tags_counter[tag] += 1  # assuming the tags are in list format, use eval() to convert string to list

# Get the 50 most common tags
top_tags = [tag for tag, _ in tags_counter.most_common(50)]

# Create a list to hold the tag and view_count information
tag_view_list = []
for i, row in data.iterrows():
    for tag in eval(row['tags']):
        if tag in top_tags:
            tag_view_list.append({'tags': tag, 'view_count': row['view_count']})

# Convert the list of dictionaries to a DataFrame
tag_view_df = pd.DataFrame(tag_view_list)

display(tag_view_df.head())
```

```python
#categorywise and view count
tag_view_df = tag_view_df.groupby('tags')['view_count'].sum().reset_index()
tag_view_df = tag_view_df.head(30)
tag_view_df.head()
```

```python
#categorywise and view count
plt.figure(figsize=(10, 6))
tag_view_df = tag_view_df.sort_values(by='view_count', ascending=False)
tag_view_df = tag_view_df.head(30)
# create a barplot using sns
sns.barplot(y= tag_view_df['tags'] , x= tag_view_df['view_count'] , legend=False)
plt.title("Tags with frequency")
plt.ylabel("Tags")
plt.xlabel("view count")
plt.show()
```

```python
tag_view_df = tag_view_df.head(30)
# Perform one-hot encoding on the 'tags' column
tag_view_df = pd.get_dummies(tag_view_df, columns=['tags'])

# Calculate the correlation matrix
correlation_matrix = tag_view_df.corr()

print(correlation_matrix)
```
